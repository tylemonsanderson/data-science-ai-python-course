## ğŸ“Š Setting Up the Dataset

First, let's create a realistic weather dataset. In real projects, you'd load this from a CSV file, but we'll generate it here for learning purposes.

````python
# Run this cell first to set up the data and imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import random

# Set random seed for reproducible results
np.random.seed(42)
random.seed(42)

print("ğŸŒ¤ï¸ Setting up weather data analysis project...")
print("Libraries loaded successfully!")
````

````python
# Create realistic weather data for 5 cities over 12 months
cities = ['New York', 'Los Angeles', 'Chicago', 'Miami', 'Seattle']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Temperature data (Fahrenheit) - realistic seasonal patterns
temp_data = {
    'New York': [32, 35, 45, 55, 65, 75, 80, 78, 70, 60, 50, 38],
    'Los Angeles': [65, 67, 70, 73, 75, 78, 82, 83, 80, 76, 70, 66],
    'Chicago': [28, 32, 42, 55, 66, 76, 81, 79, 71, 59, 45, 32],
    'Miami': [75, 77, 80, 83, 86, 88, 90, 90, 88, 85, 80, 76],
    'Seattle': [45, 48, 52, 58, 63, 68, 73, 74, 68, 60, 52, 46]
}

# Precipitation data (inches) - realistic regional patterns  
precip_data = {
    'New York': [3.5, 3.0, 4.2, 4.1, 4.5, 4.2, 4.4, 4.2, 3.8, 3.3, 3.7, 3.8],
    'Los Angeles': [3.1, 3.3, 2.4, 0.8, 0.2, 0.1, 0.0, 0.1, 0.3, 0.7, 1.8, 2.7],
    'Chicago': [2.1, 2.0, 2.8, 3.6, 3.9, 3.7, 3.6, 3.5, 3.0, 2.4, 2.3, 2.2],
    'Miami': [2.3, 2.1, 2.9, 3.1, 6.7, 9.2, 8.9, 8.5, 9.7, 6.2, 2.7, 2.2],
    'Seattle': [5.6, 4.2, 3.8, 2.4, 1.9, 1.5, 0.7, 1.0, 1.9, 3.4, 5.7, 6.1]
}

# Create the DataFrame
weather_data = []
for city in cities:
    for i, month in enumerate(months):
        weather_data.append({
            'city': city,
            'month': month,
            'month_num': i + 1,
            'temperature': temp_data[city][i],
            'precipitation': precip_data[city][i]
        })

df = pd.DataFrame(weather_data)

print("âœ… Weather data created successfully!")
print(f"ğŸ“Š Dataset shape: {df.shape}")
print(f"ğŸ™ï¸ Cities: {', '.join(cities)}")
print(f"ğŸ“… Months: {len(months)} months of data")
print("\n" + "="*50)
print("SAMPLE DATA:")
print(df.head(10))
````

### Task 1: Data Exploration (10 minutes)
**Goal**: Understand your dataset before analysis

````python
# Task 1: Data Exploration
print("=== DATA EXPLORATION ===")

# 1. Basic dataset information
print("1. Dataset Overview:")
print(f"   - Shape: {df.shape}")
print(f"   - Columns: {list(df.columns)}")
print(f"   - Data types:\n{df.dtypes}")

# 2. Check for missing values
print(f"\n2. Missing values:")
missing_counts = df.isnull().sum()
print(missing_counts)
if missing_counts.sum() == 0:
    print("   âœ… No missing values found!")

# 3. Summary statistics
print(f"\n3. Temperature Statistics:")
print(df['temperature'].describe())

print(f"\n4. Precipitation Statistics:")
print(df['precipitation'].describe())

# 5. Unique values
print(f"\n5. Unique cities: {df['city'].unique()}")
print(f"   Records per city: {df['city'].value_counts().iloc[0]}")

# YOUR TURN: Add more exploration
# - What's the temperature range across all data?
# - Which city appears first in the dataset?
# - Calculate the overall average precipitation

print(f"\nğŸ¯ YOUR EXPLORATION:")
temp_range = df['temperature'].max() - df['temperature'].min()
print(f"   - Temperature range: {temp_range}Â°F")
print(f"   - First city in dataset: {df['city'].iloc[0]}")
print(f"   - Overall average precipitation: {df['precipitation'].mean():.2f} inches")
````

### Task 2: Temperature Analysis (15 minutes)
**Goal**: Analyze temperature patterns across cities and seasons

````python
# Task 2: Temperature Analysis
print("=== TEMPERATURE ANALYSIS ===")

# 1. Average temperature by city
print("1. Average Temperature by City:")
city_avg_temps = df.groupby('city')['temperature'].mean().sort_values(ascending=False)
for city, temp in city_avg_temps.items():
    print(f"   {city}: {temp:.1f}Â°F")

# 2. Temperature variation (standard deviation) by city
print(f"\n2. Temperature Variation by City:")
city_temp_std = df.groupby('city')['temperature'].std().sort_values(ascending=False)
for city, std in city_temp_std.items():
    print(f"   {city}: {std:.1f}Â°F standard deviation")

# 3. Cities with average temperature > 65Â°F
print(f"\n3. Warm Cities (avg > 65Â°F):")
warm_cities = city_avg_temps[city_avg_temps > 65].index.tolist()
print(f"   {', '.join(warm_cities)}")

# 4. Hottest and coldest months overall
print(f"\n4. Temperature Extremes:")
hottest_record = df.loc[df['temperature'].idxmax()]
coldest_record = df.loc[df['temperature'].idxmin()]
print(f"   Hottest: {hottest_record['temperature']}Â°F in {hottest_record['city']}, {hottest_record['month']}")
print(f"   Coldest: {coldest_record['temperature']}Â°F in {coldest_record['city']}, {coldest_record['month']}")

# 5. Seasonal analysis
print(f"\n5. Seasonal Temperature Patterns:")
seasonal_months = {
    'Winter': [12, 1, 2], 'Spring': [3, 4, 5], 
    'Summer': [6, 7, 8], 'Fall': [9, 10, 11]
}

for season, months in seasonal_months.items():
    season_data = df[df['month_num'].isin(months)]
    avg_temp = season_data['temperature'].mean()
    print(f"   {season}: {avg_temp:.1f}Â°F average")

# YOUR TURN: Try additional analysis
# - Which city has the most consistent temperature year-round?
# - What's the temperature difference between summer and winter?
most_consistent = city_temp_std.index[-1]  # Lowest std dev
print(f"\nğŸ¯ Most consistent temperature: {most_consistent} ({city_temp_std[most_consistent]:.1f}Â°F std dev)")
````

### Task 3: Precipitation Analysis (10 minutes)
**Goal**: Understand rainfall patterns and create a classification function

````python
# Task 3: Precipitation Analysis
print("=== PRECIPITATION ANALYSIS ===")

# 1. Annual precipitation by city
print("1. Annual Precipitation by City:")
city_annual_precip = df.groupby('city')['precipitation'].sum().sort_values(ascending=False)
for city, precip in city_annual_precip.items():
    print(f"   {city}: {precip:.1f} inches/year")

# 2. Wettest months across all cities
print(f"\n2. Average Precipitation by Month:")
month_avg_precip = df.groupby('month')['precipitation'].mean()
# Reorder months properly
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for month in month_order:
    if month in month_avg_precip.index:
        print(f"   {month}: {month_avg_precip[month]:.1f} inches")

# 3. Create precipitation classification function
def classify_precipitation(inches):
    """
    Classify precipitation level
    Returns: string classification
    """
    if inches < 2:
        return "Dry"
    elif inches <= 5:
        return "Moderate"
    else:
        return "Wet"

# Apply classification
df['precip_category'] = df['precipitation'].apply(classify_precipitation)

print(f"\n3. Precipitation Classification:")
precip_counts = df['precip_category'].value_counts()
for category, count in precip_counts.items():
    print(f"   {category}: {count} months")

# 4. Wettest and driest cities/months
print(f"\n4. Precipitation Extremes:")
wettest = df.loc[df['precipitation'].idxmax()]
driest = df.loc[df['precipitation'].idxmin()]
print(f"   Wettest: {wettest['precipitation']:.1f} inches in {wettest['city']}, {wettest['month']}")
print(f"   Driest: {driest['precipitation']:.1f} inches in {driest['city']}, {driest['month']}")

# YOUR TURN: Additional analysis
# - Which city has the most "Wet" months?
wet_months_by_city = df[df['precip_category'] == 'Wet'].groupby('city').size().sort_values(ascending=False)
if len(wet_months_by_city) > 0:
    wettest_city = wet_months_by_city.index[0]
    wet_count = wet_months_by_city.iloc[0]
    print(f"\nğŸ¯ City with most wet months: {wettest_city} ({wet_count} months)")
else:
    print(f"\nğŸ¯ No months classified as 'Wet' in the dataset")
````

### Task 4: Data Visualization (15 minutes)
**Goal**: Create meaningful charts to visualize weather patterns

````python
# Task 4: Data Visualization
print("=== CREATING VISUALIZATIONS ===")

# Set up the plotting area
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Weather Data Analysis Dashboard', fontsize=16, fontweight='bold')

# Plot 1: Average temperature by city (bar chart)
city_avg_temps.plot(kind='bar', ax=axes[0,0], color='skyblue', alpha=0.8)
axes[0,0].set_title('Average Temperature by City')
axes[0,0].set_ylabel('Temperature (Â°F)')
axes[0,0].set_xlabel('City')
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].grid(axis='y', alpha=0.3)

# Plot 2: Temperature trends by month for all cities (line chart)
for city in cities:
    city_data = df[df['city'] == city]
    month_temps = city_data.groupby('month')['temperature'].mean()
    # Reorder months
    month_temps = month_temps.reindex(month_order)
    axes[0,1].plot(month_temps.index, month_temps.values, marker='o', label=city, linewidth=2)

axes[0,1].set_title('Temperature Trends by Month')
axes[0,1].set_ylabel('Temperature (Â°F)')
axes[0,1].set_xlabel('Month')
axes[0,1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0,1].grid(alpha=0.3)
axes[0,1].tick_params(axis='x', rotation=45)

# Plot 3: Temperature vs Precipitation scatter plot
colors = {'New York': 'blue', 'Los Angeles': 'red', 'Chicago': 'green', 
          'Miami': 'orange', 'Seattle': 'purple'}
for city in cities:
    city_data = df[df['city'] == city]
    axes[1,0].scatter(city_data['temperature'], city_data['precipitation'], 
                     c=colors[city], label=city, alpha=0.7, s=50)

axes[1,0].set_title('Temperature vs Precipitation')
axes[1,0].set_xlabel('Temperature (Â°F)')
axes[1,0].set_ylabel('Precipitation (inches)')
axes[1,0].legend()
axes[1,0].grid(alpha=0.3)

# Plot 4: Annual precipitation by city
city_annual_precip.plot(kind='bar', ax=axes[1,1], color='lightcoral', alpha=0.8)
axes[1,1].set_title('Annual Precipitation by City')
axes[1,1].set_ylabel('Precipitation (inches/year)')
axes[1,1].set_xlabel('City')
axes[1,1].tick_params(axis='x', rotation=45)
axes[1,1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("âœ… Visualizations created successfully!")
print("\nğŸ¯ What do you see in the charts?")
print("   - Which city has the most consistent temperature year-round?")
print("   - Is there a relationship between temperature and precipitation?")
print("   - Which cities have distinct seasonal patterns?")
````

### Bonus 1: Seasonal Analysis

````python
# Bonus 1: Seasonal Analysis
print("=== BONUS: SEASONAL ANALYSIS ===")

# Define seasons more precisely
seasons = {
    'Winter': [12, 1, 2], 
    'Spring': [3, 4, 5], 
    'Summer': [6, 7, 8], 
    'Fall': [9, 10, 11]
}

# Calculate seasonal statistics for each city
seasonal_data = []
for city in cities:
    city_data = df[df['city'] == city]
    for season, months in seasons.items():
        season_city_data = city_data[city_data['month_num'].isin(months)]
        seasonal_data.append({
            'city': city,
            'season': season,
            'avg_temp': season_city_data['temperature'].mean(),
            'avg_precip': season_city_data['precipitation'].mean(),
            'temp_range': season_city_data['temperature'].max() - season_city_data['temperature'].min()
        })

seasonal_df = pd.DataFrame(seasonal_data)

# Find cities with most distinct seasons (highest temperature range)
print("Cities with most distinct seasons (largest temp range):")
for season in seasons.keys():
    season_data = seasonal_df[seasonal_df['season'] == season]
    most_variable = season_data.loc[season_data['temp_range'].idxmax()]
    print(f"   {season}: {most_variable['city']} ({most_variable['temp_range']:.0f}Â°F range)")

# YOUR TURN: Which city has the most consistent climate year-round?
city_temp_variation = []
for city in cities:
    city_seasonal = seasonal_df[seasonal_df['city'] == city]
    temp_std = city_seasonal['avg_temp'].std()
    city_temp_variation.append({'city': city, 'seasonal_variation': temp_std})

variation_df = pd.DataFrame(city_temp_variation)
most_consistent = variation_df.loc[variation_df['seasonal_variation'].idxmin()]
print(f"\nMost consistent climate: {most_consistent['city']} ({most_consistent['seasonal_variation']:.1f}Â°F seasonal variation)")
````

### Bonus 2: Climate Similarity Matrix

````python
# Bonus 2: Climate Similarity Analysis
print("=== BONUS: CLIMATE SIMILARITY MATRIX ===")

# Calculate temperature profiles for each city
city_profiles = {}
for city in cities:
    city_data = df[df['city'] == city].sort_values('month_num')
    city_profiles[city] = city_data['temperature'].values

# Calculate correlation matrix (climate similarity)
similarity_matrix = []
for city1 in cities:
    row = []
    for city2 in cities:
        if city1 == city2:
            correlation = 1.0
        else:
            correlation = np.corrcoef(city_profiles[city1], city_profiles[city2])[0, 1]
        row.append(correlation)
    similarity_matrix.append(row)

# Display similarity matrix
print("Climate Similarity Matrix (Temperature Patterns):")
print("(1.0 = identical patterns, values closer to 1.0 = more similar)")
print(f"{'City':<12}", end="")
for city in cities:
    print(f"{city:<10}", end="")
print()

for i, city1 in enumerate(cities):
    print(f"{city1:<12}", end="")
    for j, city2 in enumerate(cities):
        print(f"{similarity_matrix[i][j]:.3f}    ", end="")
    print()

# Find most similar cities
max_similarity = 0
most_similar_pair = None
for i in range(len(cities)):
    for j in range(i+1, len(cities)):
        similarity = similarity_matrix[i][j]
        if similarity > max_similarity:
            max_similarity = similarity
            most_similar_pair = (cities[i], cities[j])

print(f"\nMost similar climate patterns: {most_similar_pair[0]} and {most_similar_pair[1]}")
print(f"Similarity score: {max_similarity:.3f}")
````
